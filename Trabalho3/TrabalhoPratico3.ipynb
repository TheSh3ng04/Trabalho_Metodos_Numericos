## Exerc√≠cio 1
#### **a)** Seja F(x) = $x^2 +sen(6x)$, para $‚àí1 ‚â§ x ‚â§ 1$

```python
import numpy as np
import math

# Defini√ß√£o da fun√ß√£o f(x)
def f(x):
    return x**2 + math.sin(6 * x)

# Par√¢metros: intervalo [-1, 1] e 8 pontos equidistantes
a, b = -1, 1
num_pontos = 8
x = np.linspace(a, b, num_pontos)  # Gerar abcissas equidistantes
y = [f(xi) for xi in x]            # Calcular valores da fun√ß√£o

# Impress√£o da tabela com 6 casas decimais
print(" i |     x_i      |     f(x_i)    ")
print("-"*40)
for i in range(8):
    print(f"{i:2d} | {x[i]:+12.6f} | {y[i]:+12.6f}")
```


### 1. Contextualiza√ß√£o e Objetivo  
O item visa gerar \( n+1 = 8 \) pontos \((x_i, f(x_i))_{i=0}^7\) que definem uma parti√ß√£o equidistante de \([-1, 1]\). Esses pontos s√£o prerequisitos para a constru√ß√£o do polin√¥mio interpolador e spline c√∫bico natural nos itens 1.(b) e 1.(c), pois ambas as aproxima√ß√µes dependem de pontos conhecidos da fun√ß√£o original.  

### 2. Par√¢metros Matem√°ticos Fundamentais  
- **N√∫mero de subintervalos**: \( n = 7 \) (derivado de \( n+1 = 8 \) pontos).  
- **Passo equidistante (\( h \))**: Calculado por \( h = \frac{b - a}{n} = \frac{1 - (-1)}{7} = \frac{2}{7} \approx 0.2857 \), garantindo espa√ßamento uniforme entre abcissas.  
- **F√≥rmula das abcissas**: \( x_i = a + i \cdot h \) (progress√£o aritm√©tica), com \( a = -1 \), \( b = 1 \) e \( i = 0, 1, ..., 7 \) (ex.: \( x_0 = -1.0 \), \( x_4 \approx 0.1429 \), \( x_7 = 1.0 \)).  

### 3. Validade da Fun√ß√£o \( f(x) \)  
\( f(x) = x^2 + \sin(6x) \) √© cont√≠nua e diferenci√°vel em \([-1, 1]\) (soma de \( x^2 \) ‚Äî polin√¥mio, portanto cont√≠nuo ‚Äî e \( \sin(6x) \) ‚Äî trigonom√©trica, tamb√©m cont√≠nua). Essa propriedade evita descontinuidades que comprometam a qualidade das aproxima√ß√µes subsequentes.  

### 4. Precis√£o e Relev√¢ncia Pr√°tica  
- Os valores de \( f(x_i) \) s√£o apresentados com 6 casas decimais, equilibrando clareza e precis√£o para os c√°lculos de interpola√ß√£o.  
- O uso de `numpy.linspace` no c√≥digo garante gera√ß√£o precisa das abcissas, sem erros de arredondamento associados a c√°lculos manuais.

- 1.(b): Construir o polin√¥mio interpolador (grau \( \leq 7 \), pois passa por 8 pontos) e o spline c√∫bico natural (7 segmentos de grau 3).
